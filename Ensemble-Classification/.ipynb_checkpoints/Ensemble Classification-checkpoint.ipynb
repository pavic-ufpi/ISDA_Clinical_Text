# Ensemble Fusion/Voting

## Votação I - Hard Voting/ Majoritory Voting 

```python
predictions = list()
for i in range(len(data_train_ng_tfidf)):
    predictions.append([data_train_ng_tfidf['LABEL-PREDICTED-RF'][i], data_train_snorkel['LABEL-PREDICTED-RF'][i], data_train_bert['LABEL-PREDICTED-RF'][i],
                        data_train_ng_tfidf['LABEL-PREDICTED-MLP'][i], data_train_snorkel['LABEL-PREDICTED-MLP'][i], data_train_bert['LABEL-PREDICTED-MLP'][i],
                        data_train_ng_tfidf['LABEL-PREDICTED-SVM'][i], data_train_snorkel['LABEL-PREDICTED-SVM'][i], data_train_bert['LABEL-PREDICTED-SVM'][i]])

maj = hard_voting(predictions)
metrics(data_train_ng_tfidf['LABEL'], maj)
```

```python
data_train['LABEL-FINAL-MAJ'] = maj
errors_maj = data_train.loc[data_train['LABEL-FINAL-MAJ'] != data_train['LABEL']]
errors_maj.head()
```

```python
errors_maj.LABEL.value_counts()
```

## Votação II - Soft Voting - Average Without Weights

```python
probas_clf1_ng = [item for sublist in probas_clf1_ng for item in sublist]
probas_clf2_ng = [item for sublist in probas_clf2_ng for item in sublist]
probas_clf3_ng = [item for sublist in probas_clf3_ng for item in sublist]
probas_clf1_snorkel = [item for sublist in probas_clf1_snorkel for item in sublist]
probas_clf2_snorkel = [item for sublist in probas_clf2_snorkel for item in sublist]
probas_clf3_snorkel = [item for sublist in probas_clf3_snorkel for item in sublist]
probas_clf1_bert = [item for sublist in probas_clf1_bert for item in sublist]
probas_clf2_bert = [item for sublist in probas_clf2_bert for item in sublist]
probas_clf3_bert = [item for sublist in probas_clf3_bert for item in sublist]
```

```python
labels_combined = list()
for i in range(len(probas_clf1_ng)):
    labels_combined.append([probas_clf1_ng[i], probas_clf1_snorkel[i], probas_clf1_bert[i],
                            probas_clf2_ng[i], probas_clf2_snorkel[i], probas_clf2_bert[i],
                            probas_clf3_ng[i], probas_clf3_snorkel[i], probas_clf3_bert[i]])
    
labels_combined = np.asarray(labels_combined)
label_soft, _ = soft_voting(labels_combined)
metrics(data_train['LABEL'], label_soft)
```

```python
data_train['LABEL-FINAL-SOFT'] = label_soft
errors_soft = data_train.loc[data_train['LABEL-FINAL-SOFT'] != data_train['LABEL']]
errors_soft.head()
```

```python
errors_soft.LABEL.value_counts()
```
