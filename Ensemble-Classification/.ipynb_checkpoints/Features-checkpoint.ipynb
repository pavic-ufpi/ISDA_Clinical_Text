```python
import copy
import itertools
import numpy as np
import pandas as pd
```

```python
def get_median_num_words_per_sample(sample):
    num_words = len(sample.split())
    return np.median(num_words)
```

```python
def get_num_words_per_sample(sample):
    num_words = len(sample.split()) 
    return np.sum(num_words)
```

```python
def get_num_unique_words_per_sample(sample):
    words = sample.split()
    words = list(itertools.chain(*words))
    unique = set(words) 
    return len(unique)
```

```python
data = pd.read_csv('C:\\Users\\orran\\OneDrive\\Documentos\\GitHub\\Research-Prescriptions\\3k\\Experimento\\data\\teste\\data.csv', encoding='latin-1')
data
```

```python
data_feat = copy.copy(data[['DATA', 'TEXTO']])
data_feat['LABEL'] = ''
data_feat
```

```python
for i in range(len(data_feat)):
    #print(get_num_words_per_sample(data_feat.TEXTO[i]))
    #print(get_num_unique_words_per_sample(data_feat.TEXTO[i]))
    if get_num_words_per_sample(data_feat.TEXTO[i]) >= 32:
        data_feat.LABEL.values[i] = 0
    elif get_num_words_per_sample(data_feat.TEXTO[i]) >= 23 and get_num_words_per_sample(data_feat.TEXTO[i]) <= 32:
        data_feat.LABEL.values[i] = 1
    elif get_num_words_per_sample(data_feat.TEXTO[i]) < 23:
        data_feat.LABEL.values[i] = 2
```

```python
data_feat
```

```python
from sklearn.metrics import accuracy_score, cohen_kappa_score, f1_score, roc_auc_score, recall_score, precision_score
from sklearn.metrics import roc_curve, auc

def metrics(y_test, y_pred, y_pred_proba=None):
    acc_score = accuracy_score(y_test, y_pred)
    kappa_score = cohen_kappa_score(y_test, y_pred)
    f_score = f1_score(y_test, y_pred, average='macro')
    prec_score = precision_score(y_test, y_pred, average='macro')
    rec_score = recall_score(y_test, y_pred, average='macro')
    
    print('Acurácia média de {:.2f}'.format(acc_score))
    print('Kappa médio de {:.2f}'.format(kappa_score))
    print('F1-score médio de {:.2f}'.format(f_score))
    print('Precisão média de {:.2f}'.format(prec_score))
    print('Recall média de {:.2f}'.format(rec_score))
    
    if y_pred_proba is not None:
        auc_score = roc_auc_score(y_test, y_pred_proba, multi_class='ovr')
        print('AUC média de {:.2f}'.format(auc_score))
```

```python
y_true = data.LABEL.values
y_pred = data_feat.LABEL.values
```

```python
metrics(list(y_true), list(y_pred))
```
